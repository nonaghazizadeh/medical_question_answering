```python
#new data
artificial_articles = df_samples['context']
labeled_articles = df_labeled['context']
unlabeled_articles = df_unlabeled['context']
```

## embedding

```python
model = SentenceTransformer('pritamdeka/BioBERT-mnli-snli-scinli-scitail-mednli-stsb')
```

```python
def get_embedding(input_text):

  embedding_input = model.encode(input_text)
  embedding_input = torch.tensor(embedding_input)

  torch.save(embedding_input, 'embedding_inputs.pt')
  embedding_input = torch.load('embedding_inputs.pt')

  return embedding_input
```

data address

```python
artificial_embeddings = get_embedding(artificial_articles)
```

```python
labeled_embeddings = get_embedding(labeled_articles)
```

```python
unlabeled_embeddings = get_embedding(unlabeled_articles)
```

## Save to drive

```python
def writein_drive(embedding_file_name):

  folder_id = '1KhE81kB_D1ESBr1MZBGm0Tq4tfQ5X0Dv' #project folder id

  current_directory = os.getcwd()
  embedding_file_path = os.path.join(current_directory, embedding_file_name)

  file = drive.CreateFile({'title': embedding_file_name, 'parents': [{'id': folder_id}]})
  file.SetContentFile(embedding_file_path)
  file.Upload()

  file_id = file['id']
  print('File ID:', file_id)

  return file_id
```
